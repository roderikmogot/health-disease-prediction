```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./heart_2020_cleaned.csv'),
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
# encode object values
obj_list = data.select_dtypes(include='object').columns
obj_list
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for obj in obj_list:
    data[obj] = le.fit_transform(data[obj].astype(str))
```

```python
data.head()
```

```python
X = data.drop('HeartDisease', axis=1).values
y = data['HeartDisease'].values
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(k_neighbors=10, random_state=42)
X, y = sm.fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split

np.random.seed(42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

np.random.seed(42)

rf = RandomForestClassifier().fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# 100 estimators
rf = RandomForestClassifier(n_estimators=100).fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier().fit(X_train, y_train)
knn.score(X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression

np.random.seed(42)

lr = LogisticRegression().fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score

clf = KNeighborsClassifier()
cv_acc = cross_val_score(clf, X, y, scoring='accuracy', cv=10)
print(f"cross-val accuracy (mean): {np.mean(cv_acc)*100:.2f}%")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

sns.set(font_scale=1.5)

clf = RandomForestClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

def plot_conf_mat(y_test, y_pred):
  fig, ax = plt.subplots(figsize=(10,7))
  ax = sns.heatmap(confusion_ma`trix(y_test, y_pred), annot=True, cbar=False)
  plt.xlabel("True label")
  plt.ylabel("Predicted label")

plot_conf_mat(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

clf = KNeighborsClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
```
